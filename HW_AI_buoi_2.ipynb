```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl


khoang_cach = ctrl.Antecedent(np.arange(0, 51, 1), 'khoang_cach')
luu_luong = ctrl.Antecedent(np.arange(0, 101, 1), 'luu_luong')
nhu_cau = ctrl.Antecedent(np.arange(0, 101, 1), 'nhu_cau')
thoi_tiet = ctrl.Antecedent(np.arange(0, 11, 1), 'thoi_tiet')
danh_gia = ctrl.Antecedent(np.arange(1, 5.1, 0.1), 'danh_gia')
dung_gio = ctrl.Antecedent(np.arange(0, 101, 1), 'dung_gio')


he_so_gia = ctrl.Consequent(np.arange(1, 5.1, 0.1), 'he_so_gia')
thuong = ctrl.Consequent(np.arange(0, 11, 1), 'thuong')

khoang_cach['ngan'] = fuzz.trimf(khoang_cach.universe, [0, 0, 3])
khoang_cach['trung_binh'] = fuzz.trimf(khoang_cach.universe, [2, 5, 8])
khoang_cach['dai'] = fuzz.trimf(khoang_cach.universe, [6, 13, 20])
khoang_cach['rat_xa'] = fuzz.trimf(khoang_cach.universe, [15, 50, 50])

luu_luong['thap'] = fuzz.trimf(luu_luong.universe, [0, 0, 30])
luu_luong['trung_binh'] = fuzz.trimf(luu_luong.universe, [20, 45, 70])
luu_luong['cao'] = fuzz.trimf(luu_luong.universe, [60, 100, 100])

nhu_cau['thap'] = fuzz.trimf(nhu_cau.universe, [0, 0, 30])
nhu_cau['trung_binh'] = fuzz.trimf(nhu_cau.universe, [20, 45, 70])
nhu_cau['cao'] = fuzz.trimf(nhu_cau.universe, [60, 100, 100])

thoi_tiet['tot'] = fuzz.trimf(thoi_tiet.universe, [0, 0, 4])
thoi_tiet['trung_binh'] = fuzz.trimf(thoi_tiet.universe, [3, 5, 7])
thoi_tiet['xau'] = fuzz.trimf(thoi_tiet.universe, [6, 10, 10])

danh_gia['kem'] = fuzz.trimf(danh_gia.universe, [1.0, 1.0, 2.5])
danh_gia['trung_binh'] = fuzz.trimf(danh_gia.universe, [2.0, 3.0, 4.0])
danh_gia['tot'] = fuzz.trimf(danh_gia.universe, [3.5, 5.0, 5.0])

dung_gio['tre'] = fuzz.trimf(dung_gio.universe, [0, 0, 50])
dung_gio['dung_gio'] = fuzz.trimf(dung_gio.universe, [40, 60, 80])
dung_gio['som'] = fuzz.trimf(dung_gio.universe, [70, 100, 100])

he_so_gia['thap'] = fuzz.trimf(he_so_gia.universe, [1, 1, 2])
he_so_gia['trung_binh'] = fuzz.trimf(he_so_gia.universe, [1.5, 2.5, 3.5])
he_so_gia['cao'] = fuzz.trimf(he_so_gia.universe, [3, 4, 4.5])
he_so_gia['rat_cao'] = fuzz.trimf(he_so_gia.universe, [4, 5, 5])

thuong['khong_co'] = fuzz.trimf(thuong.universe, [0, 0, 2])
thuong['it'] = fuzz.trimf(thuong.universe, [1, 3, 5])
thuong['trung_binh'] = fuzz.trimf(thuong.universe, [4, 6, 8])
thuong['cao'] = fuzz.trimf(thuong.universe, [7, 10, 10])


rule1 = ctrl.Rule(khoang_cach['ngan'] & luu_luong['thap'] & nhu_cau['thap'], he_so_gia['thap'])
rule2 = ctrl.Rule(khoang_cach['ngan'] & luu_luong['trung_binh'] & nhu_cau['cao'], he_so_gia['trung_binh'])
rule3 = ctrl.Rule(khoang_cach['trung_binh'] & luu_luong['cao'] & nhu_cau['cao'], he_so_gia['cao'])
rule4 = ctrl.Rule(khoang_cach['dai'] & luu_luong['trung_binh'] & thoi_tiet['tot'], he_so_gia['trung_binh'])
rule5 = ctrl.Rule(khoang_cach['dai'] & luu_luong['cao'] & thoi_tiet['xau'], he_so_gia['rat_cao'])
rule6 = ctrl.Rule(khoang_cach['rat_xa'] & luu_luong['cao'] & nhu_cau['cao'], he_so_gia['rat_cao'])
rule7 = ctrl.Rule(khoang_cach['trung_binh'] & luu_luong['thap'] & nhu_cau['thap'], he_so_gia['trung_binh'])
rule8 = ctrl.Rule(khoang_cach['ngan'] & luu_luong['cao'] & thoi_tiet['xau'], he_so_gia['cao'])
rule9 = ctrl.Rule(khoang_cach['rat_xa'] & thoi_tiet['xau'], he_so_gia['rat_cao'])
rule10 = ctrl.Rule(khoang_cach['trung_binh'] & luu_luong['trung_binh'] & thoi_tiet['trung_binh'], he_so_gia['trung_binh'])

rule11 = ctrl.Rule(danh_gia['tot'] & dung_gio['som'], thuong['cao'])
rule12 = ctrl.Rule(danh_gia['trung_binh'] & dung_gio['dung_gio'], thuong['trung_binh'])
rule13 = ctrl.Rule(danh_gia['kem'] & dung_gio['tre'], thuong['khong_co'])
rule14 = ctrl.Rule(khoang_cach['dai'] & luu_luong['cao'] & dung_gio['dung_gio'], thuong['cao'])
rule15 = ctrl.Rule(khoang_cach['trung_binh'] & luu_luong['trung_binh'] & danh_gia['tot'], thuong['trung_binh'])
rule16 = ctrl.Rule(danh_gia['kem'] & dung_gio['tre'], thuong['khong_co'])
rule17 = ctrl.Rule(khoang_cach['rat_xa'] & thoi_tiet['xau'] & danh_gia['tot'], thuong['cao'])
rule18 = ctrl.Rule(khoang_cach['ngan'] & danh_gia['trung_binh'] & dung_gio['dung_gio'], thuong['it'])
rule19 = ctrl.Rule(khoang_cach['dai'] & luu_luong['cao'] & dung_gio['tre'], thuong['it'])
rule20 = ctrl.Rule(khoang_cach['trung_binh'] & thoi_tiet['trung_binh'] & danh_gia['tot'], thuong['trung_binh'])

# --- CÁC LUẬT BỔ SUNG ĐỂ TRÁNH LỖI KHI NHẬP SỐ NGẪU NHIÊN ---
rule21 = ctrl.Rule(khoang_cach['dai'] & luu_luong['trung_binh'] & thoi_tiet['trung_binh'], he_so_gia['trung_binh'])
rule22 = ctrl.Rule(danh_gia['tot'] & dung_gio['tre'], thuong['khong_co'])

grabbike_ctrl = ctrl.ControlSystem([
    rule1, rule2, rule3, rule4, rule5, rule6, rule7, rule8, rule9, rule10,
    rule11, rule12, rule13, rule14, rule15, rule16, rule17, rule18, rule19, rule20,
    rule21, rule22
])
grabbike_sim = ctrl.ControlSystemSimulation(grabbike_ctrl)


DON_GIA_CO_BAN = 6000


input_khoang_cach = 10  #
grabbike_sim.input['khoang_cach'] = input_khoang_cach
grabbike_sim.input['luu_luong'] = 50
grabbike_sim.input['nhu_cau'] = 80
grabbike_sim.input['thoi_tiet'] = 5
grabbike_sim.input['danh_gia'] = 4.0
grabbike_sim.input['dung_gio'] = 0


grabbike_sim.compute()


he_so_tra_ve = grabbike_sim.output['he_so_gia']
gia_goc = input_khoang_cach * DON_GIA_CO_BAN
gia_cuoi_cung = gia_goc * he_so_tra_ve
diem_thuong = grabbike_sim.output['thuong']


print("="*40)
print("     HÓA ĐƠN ƯỚC TÍNH GRABBIKE")
print("="*40)
print(f"Quãng đường di chuyển  : {input_khoang_cach} km")
print(f"Đơn giá cơ bản         : {DON_GIA_CO_BAN:,.0f} VNĐ/km")
print(f"-> GIÁ GỐC             : {gia_goc:,.0f} VNĐ")
print("-" * 40)
print(f"Hệ số giá (Fuzzy Logic): {he_so_tra_ve:.2f}x (Do kẹt xe/thời tiết)")
print(f"-> GIÁ CƯỚC THỰC TẾ    : {gia_cuoi_cung:,.0f} VNĐ")
print("-" * 40)
print(f"Điểm thưởng cho tài xế : {diem_thuong:.2f} điểm / 10")
print("="*40)
```

```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl




mat_do = ctrl.Antecedent(np.arange(0, 101, 1), 'mat_do')
khan_cap = ctrl.Antecedent(np.arange(0, 101, 1), 'khan_cap')
tai_trong = ctrl.Antecedent(np.arange(0, 101, 1), 'tai_trong')
giao_thong = ctrl.Antecedent(np.arange(0, 61, 1), 'giao_thong')
loi_nhuan = ctrl.Antecedent(np.arange(0, 101, 1), 'loi_nhuan')


ket_hop = ctrl.Consequent(np.arange(1, 11, 1), 'ket_hop')

uu_tien = ctrl.Consequent(np.arange(0, 11, 1), 'uu_tien')



mat_do['thap'] = fuzz.trimf(mat_do.universe, [0, 0, 45])
mat_do['trung_binh'] = fuzz.trimf(mat_do.universe, [35, 55, 75])
mat_do['cao'] = fuzz.trimf(mat_do.universe, [65, 100, 100])


khan_cap['thap'] = fuzz.trimf(khan_cap.universe, [0, 0, 45])
khan_cap['trung_binh'] = fuzz.trimf(khan_cap.universe, [35, 55, 75])
khan_cap['cao'] = fuzz.trimf(khan_cap.universe, [65, 100, 100])

tai_trong['thap'] = fuzz.trimf(tai_trong.universe, [0, 0, 45])
tai_trong['trung_binh'] = fuzz.trimf(tai_trong.universe, [35, 55, 75])
tai_trong['cao'] = fuzz.trimf(tai_trong.universe, [65, 100, 100])


giao_thong['cao'] = fuzz.trimf(giao_thong.universe, [0, 0, 25])
giao_thong['trung_binh'] = fuzz.trimf(giao_thong.universe, [15, 30, 45])
giao_thong['thap'] = fuzz.trimf(giao_thong.universe, [35, 60, 60])


loi_nhuan['thap'] = fuzz.trimf(loi_nhuan.universe, [0, 0, 45])
loi_nhuan['trung_binh'] = fuzz.trimf(loi_nhuan.universe, [35, 55, 75])
loi_nhuan['cao'] = fuzz.trimf(loi_nhuan.universe, [65, 100, 100])


ket_hop['it'] = fuzz.trimf(ket_hop.universe, [1, 1, 4])
ket_hop['mot_so'] = fuzz.trimf(ket_hop.universe, [3, 5, 7])
ket_hop['nhieu'] = fuzz.trimf(ket_hop.universe, [6, 10, 10])

uu_tien['thap'] = fuzz.trimf(uu_tien.universe, [0, 0, 4])
uu_tien['trung_binh'] = fuzz.trimf(uu_tien.universe, [3, 5, 7])
uu_tien['cao'] = fuzz.trimf(uu_tien.universe, [6, 10, 10])




r1 = ctrl.Rule(mat_do['cao'] & tai_trong['thap'] & giao_thong['thap'], ket_hop['nhieu'])
r2 = ctrl.Rule(mat_do['trung_binh'] & giao_thong['cao'] & khan_cap['trung_binh'], ket_hop['mot_so'])
r3 = ctrl.Rule(tai_trong['cao'] & mat_do['cao'] & loi_nhuan['trung_binh'], ket_hop['mot_so'])
r4 = ctrl.Rule(mat_do['thap'] & khan_cap['cao'] & giao_thong['trung_binh'], ket_hop['mot_so'])
r5 = ctrl.Rule(loi_nhuan['cao'] & khan_cap['cao'] & giao_thong['cao'], ket_hop['mot_so'])

r6 = ctrl.Rule(khan_cap['cao'] & loi_nhuan['cao'], uu_tien['cao'])
r7 = ctrl.Rule(khan_cap['trung_binh'] & giao_thong['trung_binh'], uu_tien['trung_binh'])
r8 = ctrl.Rule(khan_cap['thap'] & mat_do['cao'] & loi_nhuan['thap'], uu_tien['thap'])

logistics_ctrl = ctrl.ControlSystem([r1, r2, r3, r4, r5, r6, r7, r8])
logistics_sim = ctrl.ControlSystemSimulation(logistics_ctrl)




val_mat_do = 85
val_khan_cap = 50
val_tai_trong = 15
val_giao_thong = 40
val_loi_nhuan = 55

logistics_sim.input['mat_do'] = val_mat_do
logistics_sim.input['khan_cap'] = val_khan_cap
logistics_sim.input['tai_trong'] = val_tai_trong
logistics_sim.input['giao_thong'] = val_giao_thong
logistics_sim.input['loi_nhuan'] = val_loi_nhuan


logistics_sim.compute()


res_ket_hop = logistics_sim.output['ket_hop']
res_uu_tien = logistics_sim.output['uu_tien']


print("="*60)
print("     HỆ THỐNG TỐI ƯU HÓA GIAO HÀNG THÔNG MINH")
print("="*60)
print(f"- Mật độ đơn hàng    : {val_mat_do}% (Cao)")
print(f"- Mức độ khẩn cấp    : {val_khan_cap}% (Trung bình)")
print(f"- Tải trọng tài xế   : {val_tai_trong}% (Thấp)")
print(f"- Tốc độ giao thông  : {val_giao_thong} km/h (Trung bình)")
print(f"- Lợi nhuận dự kiến  : {val_loi_nhuan}% (Trung bình)")
print("-" * 60)
print(f"-> SỐ ĐƠN HÀNG NÊN KẾT HỢP : {int(np.round(res_ket_hop))} đơn")
print(f"-> MỨC ĐỘ ƯU TIÊN GIAO HÀNG: {res_uu_tien:.2f} / 10")
print("-" * 60)
print("Kết quả: Hệ thống đề xuất gom thêm đơn để tối ưu thu nhập")
print("nhờ mật độ cao và tài xế đang rảnh (tải trọng thấp).")
print("="*60)
```
